```python
for x1 in soup1:
    url_=x1.find(attrs={'class':'image'}).find('a')['href']
    print(url_)
    name = x1.find(attrs={'class':'product_name'}).text.strip()#.replace('\n','')
    print(name)
    try:
        price_n = x1.find(attrs={'class':'product_price'}).find(attrs={'class':'old_price'}).text.strip()
    except:
        price_n = x1.find(attrs={'class':'product_price'}).text.strip()
    try:
        price_o = x1.find(attrs={'class':'product_price'}).find(attrs={'class':'price5'}).text.strip().replace('\t','').replace('\r','').replace('\n','')
    except:
        price_o = ''
    if price_o != '':
        discount = int(str(round((1-(stringtofloat(price_o)/stringtofloat(price_n)))*100)))
    else:
        discount = 0
        
    #list_pricecmr.append(price_cmr)
    list_discount.append(discount)
    list_priceoffer.append(price_o)
    list_pricenormal.append(price_n)
    list_urls.append(url_)
    list_name.append(name)
```
